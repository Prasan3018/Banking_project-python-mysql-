```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Connect to server
cnx = mysql.connector.connect(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="admin")
```

```python
query = "select * from banking_case.customer "
```

```python
df = pd.read_sql(query, cnx)
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
# examining the distribution of unique categories in categorical columns
categorical_cols = df[["BRId","GenderId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}' :")
    display(df[col].value_counts())
```

```python
#univariate analysis
for i, predictor in enumerate(df[["BRId","GenderId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

```python
#bivariate analysis
for i, predictor in enumerate(df[["BRId","GenderId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue="GenderId",palette="Set1")
```

```python
# Histplot of value counts for different Occupation

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
# Numerical analysis
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

#univariate analysis and visualization
plt.figure(figsize=(15,25))
for i, col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

```python
# Heatmaps

numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
